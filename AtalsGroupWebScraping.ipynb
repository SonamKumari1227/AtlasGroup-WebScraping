```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

driver = webdriver.Chrome()

location_elements = []

base_url = 'https://altus.search-prop.com/properties?page={}'

for page_number in range(1, 28):
    print("Page:", page_number)
    url = base_url.format(page_number)
    print("URL:", url)
    driver.get(url)

    try:
  
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.XPATH, "//p[@class='propText__town']"))
        )

        elements = driver.find_elements(By.XPATH, "//p[@class='propText__town']")
        
        if len(elements) > 0:
            for element in elements:
                location_elements.append(element.text)
                print(element.text)
        else:
            print("No elements found on this page.")
    except TimeoutException:
        print("Timed out waiting for page to load.")


df['Location'] = location_elements        

driver.quit()


```
